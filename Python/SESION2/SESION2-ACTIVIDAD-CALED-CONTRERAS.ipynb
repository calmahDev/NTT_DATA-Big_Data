escribe un programa en python que lea el archivo datos_csv.csv del asiguiente manera
para cada fila, calcula el promedio de las tre ultimas columnas (columna -3)
si el promedio es mayor a 10, escribe la fila en un archivo nuevo llamado aprobados.csv , si el promedioo es menor a 10, escribe la fila en un archivo nuevo llamado reprobados.csv

```python
import pandas as pd
```

```python
# Nombre del archivo de entrada
df = pd.read_csv("datos_csv.csv")
```

```python
print(df)
```

```python
#Calcular el promedio en una columna adicional
df['Promedio'] = df.iloc[:, -3:].mean(axis=1)
```

```python
df
```

```python
#Hacer el filtrado de aprobados y reprobados
aprobados = df[df['Promedio'] > 10]
reprobados = df[df['Promedio'] <= 10]
```

```python
#Guardar el datafrmae filtrada en sus respectivos archivos
archivo_aprobados = 'aprobados.csv'
archivo_reprobados = 'reprobados.csv'
```

```python
# Guarda los DataFrames en archivos CSV
aprobados.to_csv(archivo_aprobados, index=False)
reprobados.to_csv(archivo_reprobados, index=False)
```

```python
#revisar el contenido de los archivos creados
aprobados = pd.read_csv("aprobados.csv")
reprobados = pd.read_csv("reprobados.csv")
```

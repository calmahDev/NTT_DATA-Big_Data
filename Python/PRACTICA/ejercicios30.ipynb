```python
import pandas as pd
```

```python
data=pd.read_csv('VENTAS_FACTURAS.csv')
data
```

```python
#Encuentra el país con el menor monto en el año 2020.
data['Monto']=data['Monto'].str.replace(',', '').astype(float)
data.dtypes
```

```python
data2020= data[data['Año']==2020]
data2020
```

```python
data_pais = data2020.groupby('Pais').sum()
data_pais
```

```python
data_pais['Monto'].idxmin()
```

```python
#Agrupa los datos por mes y calcula el total de una columna para cada mes.
```

```python
data_mes = data.groupby('Mes').sum()
data_mes['Mes' and 'Monto']
```

```python
#Convierte la columna de fechas a objetos de fecha y encuentra el día con el mayor número de eventos.
data['Fecha de factura']=pd.to_datetime(data['Fecha de factura'])
data['Dia']= data['Fecha de factura'].dt.day.astype('int64')
data['Fecha de factura'] = data['Fecha de factura'].astype(str)
data_day = data.groupby('Dia')['Monto'].sum()
data_day.idxmax() , data_day.max()
```
